```python
import warnings

import matplotlib.pyplot as plt
import networkx as nx

from dag_visualization import draw


def read_graph(filename):
    # Sometimes networkx.read_graphml produces numerous warnings,
    # but resulting graph is still very much intact.
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        graph = nx.read_graphml(filename)
    
    return graph


def plot_both_methods(filename, *, max_width=3, figsize=(16, 16)):
    graph = read_graph(filename)
    fig, (graham_coffman_ax, dummy_vertices_ax) = plt.subplots(ncols=2, figsize=figsize)

    graham_coffman_ax.set_title("Graham-Coffman layering")
    draw(graph, max_width, ax=graham_coffman_ax)

    dummy_vertices_ax.set_title("Dummy vertices minimization layering")
    draw(graph, max_width=None, ax=dummy_vertices_ax)

    plt.show()
```

```python
plot_both_methods("example1.xml", max_width=3, figsize=(16, 16))
```

```python
plot_both_methods("example2.xml", max_width=4, figsize=(16, 16))
```
