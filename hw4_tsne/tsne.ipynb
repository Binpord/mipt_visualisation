```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits

from tsne import tsne


# We will try to visualize the MNIST dataset.
digits = load_digits()
data = digits.images.reshape((len(digits.images), -1))
target = digits.target

# Run our t-SNE to reduce dimensions.
tsne_output = tsne(data, perplexity=5, gradient_descent_learning_rate=100)

# Plot the result.
plt.scatter(tsne_output[:, 0], tsne_output[:, 1], c=target)
plt.show()
```

```python
from sklearn.manifold import TSNE


# Run sklearn t-SNE in order to compare results.
tsne_output = TSNE(perplexity=5).fit_transform(data)

# Plot the result.
plt.scatter(tsne_output[:, 0], tsne_output[:, 1], c=target)
plt.show()
```
